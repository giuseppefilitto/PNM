# SELF DIFF COEFF FIT

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

## H2O

```python
def fit_func(x, a, b):
    return a*x + b
```

```python
T2obs = np.array([81.19, 43.54, 24.04, 10.98, 5.748, 2.147])
Robs = 1/T2obs

tau = np.array([25, 35, 50, 75, 100, 175])
tau = tau * 1e-3
```

```python
fit_params = curve_fit(fit_func, tau**2, Robs)
popt, pcov = fit_params
print(popt, pcov)
```

```python
print("a =", popt[0], "+/-", pcov[0,0]**0.5)
print("b =", popt[1], "+/-", pcov[1,1]**0.5)
```

```python
residuals = Robs - fit_func(tau**2, *popt)
ss_res = np.sum(residuals**2)
ss_tot = np.sum((Robs-np.mean(Robs))**2)

r_squared = 1 - (ss_res / ss_tot)

print('R2 =', r_squared)
```

```python
plt.style.use('seaborn-talk')

y  = fit_func(tau**2, popt[0], popt[1])

plt.figure(figsize=(8,8))
plt.scatter(tau**2, Robs, marker='s', s=25, label= 'data')
plt.plot(tau**2, y, 'r-', label = 'fit')
plt.legend()
plt.ylabel(r'$R_2$ ($ms^-1$)')
plt.xlabel(r'$\tau^2 (ms^2)$')
plt.title(r' EDTA-$H_2O$ ')
plt.text(0.0001, 0.4, '$R_2$' + f' = {popt[0]: .3f} x + {popt[1]: .3f}', style='italic', fontsize=16,
        bbox={'facecolor': 'b', 'alpha': 0.1, 'pad': 10})
plt.text(0.0001, 0.35, f'$R^2$ = {r_squared: .3f} ', style='italic', fontsize=16,
        bbox={'facecolor': 'g', 'alpha': 0.1, 'pad': 10})
plt.savefig('plots/R2-H2O.pdf')
```
