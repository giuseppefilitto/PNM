## H20

```python
paths_dat = glob.glob('Data/H2O/*.dat')
print(paths_dat)
```

```python
for dat in paths_dat:
    df = pd.read_csv(dat, header=None, sep=',' , engine='python') 
    df.to_csv(os.path.splitext(dat)[0] + '.csv')
    df = pd.read_csv(os.path.splitext(dat)[0] + '.csv', header=1, index_col=0)
    df.columns  = df.columns.str.replace(' ', '')
    for i, col in enumerate(df.columns):
        if df[col].dtype == 'object':
            df[col] = pd.to_numeric(df[col], errors='coerce')
    df.to_csv(os.path.splitext(dat)[0] + '.csv', index=False)
    
```

```python
dataframes = glob.glob('Data/H2O/*.csv')

print(dataframes)

sorted_list = sorted(dataframes, key=lambda x: int(os.path.split(os.path.splitext(x)[0])[1].replace('H2Oedta_edly', '').replace('1F', '')))
    
print(sorted_list)
```

```python
plt.style.use('seaborn-talk')
plt.figure(figsize=(10,8))
for k, data in enumerate(sorted_list):
    df = pd.read_csv(data)
    T2 = df['T'].values
    S = df['Sig_Np'].values
    path = os.path.splitext(data)[0]
    filename = os.path.split(path)[1]
    tau = filename.replace('H2Oedta_edly', '')
    tau = tau.replace(tau[-2::], '')
    plt.plot(T2, S, label = r'$\tau$ = ' + str(tau) + r' $\mu s$')


plt.xscale('log')
plt.ylim([0, 1000])
plt.xlim([None, 1e3])
plt.xlabel('$T_2$ (ms)')
plt.ylabel('Signal Density (a. u.)')
plt.title('EDTA-$H_2O$')
plt.legend(loc='upper left')
plt.savefig('T2-H2O.pdf')
plt.show()

```

```python
plt.style.use('seaborn-talk')
plt.figure(figsize=(10,8))

for k, data in enumerate(sorted_list):
    df = pd.read_csv(data)
    t = df['SigT'].values
    S = df['Sig'].values
    fit = df['SigCalc']
    path = os.path.splitext(data)[0]
    filename = os.path.split(path)[1]
    tau = filename.replace('H2Oedta_edly', '')
    tau = tau.replace(tau[-2::], '')
    plt.scatter(t, S, marker='s', s=10, label =  r'$\tau$ = ' + str(tau) + r' $\mu s$')
    plt.plot(t, fit, lw=1, label = r'$\tau$ = ' + str(tau) + r' $\mu s$ ')

    
    h, l = plt.gca().get_legend_handles_labels()

    plt.legend(handles=zip(h[k+1::], h[::-k+len(sorted_list)]), labels=l, handler_map = {tuple: HandlerTuple(None)})

plt.yscale('log')
plt.xlabel('t (ms)')
plt.ylabel('NMR Signal (a. u.)')
plt.title('EDTA-$H_2O$')
plt.ylim([1e-2, 1e3])
plt.savefig('Signal-H2O.pdf')
plt.show()
```

## BULK WATER AND VICENZA STONE

```python
paths_dat = glob.glob('Data/Bulk_WATER_and_Vicenza/*.dat')
print(paths_dat)

```

```python
for dat in paths_dat:
    df = pd.read_csv(dat, header=None, sep=',' , engine='python') 
    df.to_csv(os.path.splitext(dat)[0] + '.csv')
    df = pd.read_csv(os.path.splitext(dat)[0] + '.csv', header=1, index_col=0)
    df.columns  = df.columns.str.replace(' ', '')
    for i, col in enumerate(df.columns):
        if df[col].dtype == 'object':
            df[col] = pd.to_numeric(df[col], errors='coerce')
    df.to_csv(os.path.splitext(dat)[0] + '.csv', index=False)
    
```

```python
dataframes = glob.glob('Data/Bulk_WATER_and_Vicenza/*.csv')

print(dataframes)
```

```python
plt.style.use('seaborn-talk')

plt.figure(figsize=(10,8))
for k, data in enumerate(dataframes):
    df = pd.read_csv(data)
    T2 = df['T'].values
    S = df['Sig_Np'].values
    path = os.path.splitext(data)[0]
    filename = os.path.split(path)[1]
    
    if filename.endswith('Vicenza_CPMG'):
        
        name = filename.replace('Vicenza_CPMG', 'Vicenza Stone')
    
    else:
        name = filename.replace('bulk_water_CPMG', 'Bulk Water')
    
              
    plt.plot(T2, S, label = name)

plt.xscale('log')


plt.ylim([0, 6e5])
plt.xlim([1, 1e4])
plt.xlabel('$T_2$ (ms)')
plt.ylabel('Signal Density (a. u.)')
plt.title('Bulk Water and Saturated Porous Medium')
plt.legend(loc='best')
plt.savefig('SurfaceEffect.pdf')
plt.show()
```

```python
plt.style.use('seaborn-talk')
plt.figure(figsize=(10,8))

for k, data in enumerate(dataframes):
    df = pd.read_csv(data)
    t = df['SigT'].values
    S = df['Sig'].values
    fit = df['SigCalc']
    path = os.path.splitext(data)[0]
    filename = os.path.split(path)[1]
    
    if filename.endswith('Vicenza_CPMG'):
        
        name = filename.replace('Vicenza_CPMG', 'Vicenza Stone')
    
    else:
        name = filename.replace('bulk_water_CPMG', 'Bulk Water')

    plt.scatter(t, S, marker='s', s=10, label =  name)
    plt.plot(t, fit, lw=1, label = name)

    
    h, l = plt.gca().get_legend_handles_labels()

    plt.legend(handles=zip(h[k+1::], h[::-k+len(dataframes)]), labels=l, handler_map = {tuple: HandlerTuple(None)})

plt.yscale('log')
plt.xlabel('t (ms)')
plt.ylabel('NMR Signal (a. u.)')
plt.title('NMR Data and Fit')
plt.ylim([1e3, None])
plt.savefig('SurfaceEffect2.pdf')
plt.show()
```
